### Analysis and Comparison of Sorting Algorithms
Sorting algorithms are essential in managing data efficiently, especially when it comes to searching, visualizing, or making decisions based on ordered information. Here's a deeper look into the behavior of the sorting algorithms used in this project:

<h2 style="color:green;text-align:center"> 1. Bubble Sort (Sorted by ID)</h2>
Time Complexity:

Worst Case: O(n²)

Best Case: O(n) (when the list is already sorted)

Average Case: O(n²)

Space Complexity: O(1) (in-place sorting)

Use Case Fit: Sorting student IDs, which are structured and unique, works fine with Bubble Sort for small lists.

Advantages:

Simple to understand and implement.

Can detect if the list is already sorted, allowing for early termination.

Disadvantages:

Extremely inefficient for large datasets.

Performs unnecessary comparisons and swaps.

Conclusion:

Works fine here because of the small dataset. However, in real-world applications with thousands of student records, it would be unsuitable.

```python
 
Students = [
    {'id':'DBU1500001', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.34},
    {'id':'DBU1500002', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id':'DBU1500003', 'name': 'Soliana', 'dept': 'DS', 'cgpa': 2.74},
    {'id':'DBU1500004', 'name': 'Rahel', 'dept': 'CS', 'cgpa': 3.1},
    {'id':'DBU1500005', 'name': 'Meklit', 'dept': 'IT', 'cgpa': 3.85},
    {'id':'DBU1500006', 'name': 'Adis', 'dept': 'CS', 'cgpa': 2.91},
    {'id':'DBU1500007', 'name': 'Abel', 'dept': 'DS', 'cgpa': 3.6},
    {'id':'DBU1500008', 'name': 'Surafel', 'dept': 'IT', 'cgpa': 3.4},
    {'id':'DBU1500009', 'name': 'Nahom', 'dept': 'DS', 'cgpa': 2.98},
    {'id':'DBU1500010', 'name': 'tsedey', 'dept': 'CS', 'cgpa': 3.45}
]

def bubble_sort_by_id(students):
    n = len(students)
    for i in range(n):
        for j in range(0, n - i - 1):
            if students[j]['id'] > students[j + 1]['id']:
                students[j], students[j + 1] = students[j + 1], students[j]
    return students

bubble_sorted = bubble_sort_by_id(Students.copy())
print("Bubble Sorted by ID:")
for student in bubble_sorted:
    print(student)
```

 <h2 style= "color:blue;text-align:center"> 2. Insertion Sort (Sorted by Name Alphabetically)</h2>
<h3><un>Time Complexity:</un></h3>

Worst Case: O(n²) (when input is sorted in reverse)

Best Case: O(n) (when already sorted)

Average Case: O(n²)

<h3><un>Space Complexity:</un></h3> O(1)

Use Case Fit: Sorting names alphabetically is a natural use case for insertion sort, especially in small to medium-sized lists.

<h3><un>Advantages:</un></h3>

More efficient than Bubble Sort on nearly sorted data.

Very efficient for small datasets.

Stable sort (preserves the relative order of records with equal keys).

<h3><un>Disadvantages:</un></h3>

Poor performance on large datasets.

Conclusion:

A good choice when the list is expected to be nearly sorted or small, which fits our student list scenario.



```python
 
Students = [
    {'id':'DBU1500001', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.34},
    {'id':'DBU1500002', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id':'DBU1500003', 'name': 'Soliana', 'dept': 'DS', 'cgpa': 2.74},
    {'id':'DBU1500004', 'name': 'Rahel', 'dept': 'CS', 'cgpa': 3.1},
    {'id':'DBU1500005', 'name': 'Meklit', 'dept': 'IT', 'cgpa': 3.85},
    {'id':'DBU1500006', 'name': 'Adis', 'dept': 'CS', 'cgpa': 2.91},
    {'id':'DBU1500007', 'name': 'Abel', 'dept': 'DS', 'cgpa': 3.6},
    {'id':'DBU1500008', 'name': 'Surafel', 'dept': 'IT', 'cgpa': 3.4},
    {'id':'DBU1500009', 'name': 'Nahom', 'dept': 'DS', 'cgpa': 2.98},
    {'id':'DBU1500010', 'name': 'tsedey', 'dept': 'CS', 'cgpa': 3.45}
]

def insertion_sort_by_name(students):
    for i in range(1, len(students)):
        key = students[i]
        j = i - 1
        while j >= 0 and key['name'].lower() < students[j]['name'].lower():
            students[j + 1] = students[j]
            j -= 1
        students[j + 1] = key
    return students

insertion_sorted = insertion_sort_by_name(Students.copy())
print("\nInsertion Sorted by Name:")
for student in insertion_sorted:
    print(student)
```

  <h2 style=" color:red;text-align: center"> 3. Selection Sort (Sorted by CGPA - Descending)</h2>
<h3><un>Time Complexity:</un></h3>
Worst, Best, and Average Case: O(n²)

<h3><un>Space Complexity: O(1)</un></h3>

Use Case Fit: Sorting by CGPA from highest to lowest is effectively done by selecting the maximum at each pass.

<h3><un>Advantages:</un></h3>

Simple to implement.

Performs well when data movement is costly (as it minimizes swaps).

<h3><un>Disadvantages:</un></h3>

Still inefficient compared to modern algorithms like Merge Sort or Quick Sort.

Doesn’t adapt to already sorted input.

Conclusion:

Practical for very small datasets, but not optimal for performance-heavy systems like academic portals or large databases.

```python
 
Students = [
    {'id':'DBU1500001', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.34},
    {'id':'DBU1500002', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id':'DBU1500003', 'name': 'Soliana', 'dept': 'DS', 'cgpa': 2.74},
    {'id':'DBU1500004', 'name': 'Rahel', 'dept': 'CS', 'cgpa': 3.1},
    {'id':'DBU1500005', 'name': 'Meklit', 'dept': 'IT', 'cgpa': 3.85},
    {'id':'DBU1500006', 'name': 'Adis', 'dept': 'CS', 'cgpa': 2.91},
    {'id':'DBU1500007', 'name': 'Abel', 'dept': 'DS', 'cgpa': 3.6},
    {'id':'DBU1500008', 'name': 'Surafel', 'dept': 'IT', 'cgpa': 3.4},
    {'id':'DBU1500009', 'name': 'Nahom', 'dept': 'DS', 'cgpa': 2.98},
    {'id':'DBU1500010', 'name': 'tsedey', 'dept': 'CS', 'cgpa': 3.45}
]

def selection_sort_by_cgpa(students):
    n = len(students)
    for i in range(n):
        max_idx = i
        for j in range(i + 1, n):
            if students[j]['cgpa'] > students[max_idx]['cgpa']:
                max_idx = j
        students[i], students[max_idx] = students[max_idx], students[i]
    return students

selection_sorted = selection_sort_by_cgpa(Students.copy())
print("\nSelection Sorted by CGPA (Descending):")
for student in selection_sorted:
    print(student)
```
